```python
#importing libraries
import numpy as np # array manager
import matplotlib.pyplot as plt #data visulization
from sklearn.datasets import load_boston #datasets
```

```python
#get data and assign to var
boston = load_boston() #load datasets
X = np.array(boston.data[:, 5]) #get numbers of rooms
Y = np.array(boston.target) #get precio medio de las viviendas

initialX = X #cache of initial x value
```

```python
#add ones column
X = np.array([np.ones(X.shape[0]), X]) #[[ones], [X]]
X = X.T
```

```python
#minimos cuadrados ordinarios
B = np.linalg.inv(X.T @ X) @ X.T @ Y
print(X.T @ Y)
```

```python
#datos
X0 = 4
X1 = 9
```

```python
#solucion de Y = W0 + W1 * Xn
Y0 = B[0] + B[1] * X0
Y1 = B[0] + B[1] * X1
```

```python
#data visulization
plt.scatter(initialX, Y, alpha=0.3)
plt.plot([X0, X1], [Y0, Y1], c='red')
plt.scatter(X0, Y0, color="red")
plt.show()
```
