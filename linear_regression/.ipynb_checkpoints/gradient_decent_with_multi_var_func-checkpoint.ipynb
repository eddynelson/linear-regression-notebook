```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from IPython.display import SVG
```

```python
SVG('math-f(x,y).svg')
```

```python
f = lambda th: (np.sin((1/2 * th[0]**2) - (1/4 * th[1]**2) + 3) * np.cos((2*th[0]) + 1 - np.e**th[1]))
```

```python
res = 100

_X = np.linspace(-2, 2, res) #100
_Y = np.linspace(-2, 2, res) #100
_Z = np.zeros((res, res))#100x100

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = f([x, y])
```

```python
theta = np.random.rand(2) * 4 - 2

#
print(theta)
plt.contourf(_X, _Y, _Z, 100)
plt.plot(theta[0], theta[1], "o", color='white')

#
_T = np.copy(theta)
h = 0.001

lr = 0.001

grad = np.zeros(2)

for _ in range(10000):
    for ith,th in enumerate(theta):
        _T = np.copy(theta)
        
        _T[ith] = _T[ith] + h

        deriv = (f(_T) - f(theta))/h

        grad[ith] = deriv

    theta = theta - (lr * grad)
    if _ % 100 == 0:
        plt.plot(theta[0], theta[1], ".", color='red')
    
plt.plot(theta[0], theta[1], "o", color='red')
plt.colorbar()
plt.show()
```
